```python
# Feature Engineering 
# Approach: If Players_5 will be in the Team2, label 1 will be assigned, else if Players_6 present, label 2 will be assigned.
# Converting players_5 and players_6  from categorical variables to continuous inputs using pandas 
#function pd.get_dummies. 
# This variable has only two answer choices: Players_5 and Players_6. 
# It creates a new dataframe which consists of zeros and ones. 
# Here dataframe will have a one depending on the player of a particular game in this case.


FinalMatchDataAnaly = FinalMatchDataAnaly.reset_index(drop=True)
FinalMatchDataAnaly.loc[FinalMatchDataAnaly.Players_5 == FinalMatchDataAnaly.Players_5,'Players5']=1
FinalMatchDataAnaly.loc[FinalMatchDataAnaly.Players_5 == FinalMatchDataAnaly.Players_6, 'Players5']=2

FinalMatchDataAnaly = FinalMatchDataAnaly.drop(['Players5'], axis=1)
FinalMatchDataAnaly.head()
```

```python
#convert Players_5 and Players_6 from categorical variables to continous inputs 

# Get dummy variables
final = pd.get_dummies(FinalMatchDataAnaly, prefix=['Players_1','Players_2','Players_3','Players_4','Players_6','Players_7',
                    'Players_8','Players_9','Players_10','Players_11','Players_12','Players_13','Players_14','Players_15',
                    'Players_16','Players_17','Players_18','Players_19','Players_20','Players_21','Players_22'], columns=['Players_1','Players_2','Players_3','Players_4','Players_6','Players_7',
                    'Players_8','Players_9','Players_10','Players_11','Players_12','Players_13','Players_14','Players_15',
                    'Players_16','Players_17','Players_18','Players_19','Players_20','Players_21','Players_22'])
# Separate X and y sets
X = final.drop(['Players_5'], axis=1)
y = final["Players_5"]

# Separate train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

## 5. Final Model Prediction

```python
# List of tuples before 
semi = [('Soumya Sarkar','BR Dunk')]
```

```python
def clean_and_predict(matches, Strike, final, logreg):

# Initialization of auxiliary list for data cleaning
    positions = []

 # Loop to retrieve each players position according to Strike Rate
    for match in matches:
        positions.append(Strike.loc[Strike['Player'] == match[0],'Strike_Rate'].iloc[0])
        positions.append(Strike.loc[Strike['Player'] == match[1],'Strike_Rate'].iloc[0])
    
    # Creating the DataFrame for prediction
    pred_set = []

    # Initializing iterators for while loop
    i = 0
    j = 0

    # 'i' will be the iterator for the 'positions' list, and 'j' for the list of Players (list of tuples)
    while i < len(positions):
        dict1 = {}

        # If position of first player is better then this player will be the 'Players_5', and vice-versa-versa
        if positions[i] < positions[i + 1]:
            dict1.update({'Players_5': matches[j][0], 'Players_6': matches[j][1]})
        else:
            dict1.update({'Players_5': matches[j][1], 'Players_6': matches[j][0]})

               # Append updated dictionary to the list, that will later be converted into a DataFrame
        pred_set.append(dict1)
        i += 2
        j += 1
        
        # Convert list into DataFrame
    pred_set = pd.DataFrame(pred_set)
    backup_pred_set = pred_set

    # Get dummy variables and drop Player column
    pred_set = pd.get_dummies(pred_set, prefix=['Players_5', 'Players_6'], columns=['Players_5', 'Players_6'])
    # Add missing columns compared to the model's training dataset
    missing_cols2 = set(final.columns) - set(pred_set.columns)
    for c in missing_cols2:
        pred_set[c] = 0
    pred_set = pred_set[final.columns]

    pred_set = pred_set.drop(['Players_5'], axis=1)

    # Predict!
    predictions = logreg.predict(pred_set)
    for i in range(len(pred_set)):
        print(backup_pred_set.iloc[i, 1] + " and " + backup_pred_set.iloc[i, 0])
        if predictions[i] == 1:
            print("Players_5r: " + backup_pred_set.iloc[i, 1])
        else:
            print("Players_5: " + backup_pred_set.iloc[i, 0])
        print("")
        
       
```

```python
clean_and_predict(semi, Strike, final, rf)
```
